# Useing ML model to predict whether or not a tumor is malignant or benign.

```python
#importing important lib
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import sklearn 
```

```python
#importing data set from sklearn 
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
```

```python
#geting to know the data 
```

```python
#0 represents malignant tumors and 1 represents benign tumors.
```

```python
df1=pd.DataFrame(data=df.data,columns=df.feature_names)
df1['target']=pd.Series(df.target)
```

```python
df1.head(10)
```

```python
df1['target'].value_counts()
```

```python
''' 357 tumers are benign and 212 are malignant'''
```

```python
##making the model
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=df1.drop("target",axis=1)
y=df1['target']
```

```python
y
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=34)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
#geting the accuracy 
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction)
```

```python
#trying with Nive Bayes
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model1=GaussianNB()
model1.fit(x_train,y_train)
```

```python
pre=model1.predict(x_test)
```

```python
accuracy_score(y_test,pre)
```
