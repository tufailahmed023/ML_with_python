```python
def remove_outliers(col):
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3-(1.5*IQR)
    return lower_range,upper_range
```

```python
#ApplicantIncome CoapplicantIncome LoanAmount
```

```python
lAi,uAi=remove_outliers(df["ApplicantIncome"])
df["ApplicantIncome"]=np.where(df["ApplicantIncome"]>uAi,uAi,df["ApplicantIncome"])
df["ApplicantIncome"]=np.where(df["ApplicantIncome"]<lAi,lAi,df["ApplicantIncome"])
```

```python
df.boxplot(column="ApplicantIncome")
```

```python
lAi,uAi=remove_outliers(df["CoapplicantIncome"])

df["CoapplicantIncome"]=np.where(df["CoapplicantIncome"]>uAi,uAi,df["CoapplicantIncome"])
df["CoapplicantIncome"]=np.where(df["CoapplicantIncome"]<lAi,lAi,df["CoapplicantIncome"])
```

```python
lAi,uAi=remove_outliers(df["LoanAmount"])

df["LoanAmount"]=np.where(df["LoanAmount"]>uAi,uAi,df["LoanAmount"])
df["LoanAmount"]=np.where(df["LoanAmount"]<lAi,lAi,df["LoanAmount"])
```
