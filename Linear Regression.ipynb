```python
#droping car name 
df1=df1.drop("car name",axis=1)
```

```python
df1
```

```python
#replacing 1,2,3 in orign with actual values
df1['origin']=df1['origin'].replace({1:'america',2:"europ",3:"asia"})
df1.sample(10)
```

```python
#one hot encoading
df1=pd.get_dummies(df1,columns=['origin'])
```

```python
df1.sample()
```

```python
df1.describe()
```

```python
df1.describe(include='all')
```

```python
df1.info()
```

```python
hpisDigit=pd.DataFrame(df1.horsepower.str.isdigit())
```

```python
df1[hpisDigit['horsepower']==False]
```

```python
# missing value ?
#replacing with Nan
df1=df1.replace("?",np.nan)
```

```python
df1[hpisDigit['horsepower']==False]
```

```python
df1.median()
```

```python
#replacing nan with median
medianfiller=lambda x:x.fillna(x.median())
```

```python
df1=df1.apply(medianfiller,axis=0)
```

```python
df1['horsepower']=df1['horsepower'].astype('float64')
```

```python
df1.head(5)
```

```python
sns.pairplot(df1,diag_kind="kde")
```

```python
#linear model
x=df1.drop(['mpg'],axis=1)
y=df1[['mpg']]
```

```python
#split x and y for train and test data 
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
reg_model=LinearRegression()
reg_model.fit(X_train,Y_train)
```

```python
reg_model.score(X_train,Y_train)
```

```python
reg_model.score(X_test,Y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
```

```python
poly=PolynomialFeatures(degree=2,interaction_only=True)
X_train2=poly.fit_transform(X_train)
x_test2=poly.fit_transform(X_test)
```

```python
y_pred=poly_clf.predict(x_test2)
```
