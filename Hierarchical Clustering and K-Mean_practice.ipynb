```python
#K-MEAN
```

```python
from sklearn.cluster import KMeans
```

```python
from scipy.spatial.distance import cdist
meanDistortion=[]
```

```python
cluster=range(1,10)
for k in cluster :
    model=KMeans(n_clusters=k)
    model.fit(Datascaled)
    pre=model.predict(Datascaled)
    meanDistortion.append(sum(np.min(cdist(Datascaled,model.cluster_centers_,"euclidean"),axis=1))/Datascaled.shape[0])

    
plt.plot(cluster,meanDistortion,"bx-")
plt.xlabel("k")
plt.ylabel("Average Distorction ")
plt.title("Selecting K with k elbow")
```

```python
#k=4
```

```python
fit_model=KMeans(n_clusters=4)
fit_model.fit(Datascaled)
pred=fit_model.predict(Datascaled)


df["Groups"]=pred
Datascaled["Groups"]=pred
print("group assigned : /n ",pred)
df.head(15)
```
